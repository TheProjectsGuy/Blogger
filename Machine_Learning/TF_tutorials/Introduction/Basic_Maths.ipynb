## Dealing with multi dimensional tensors
In the last tutorial, we dealt with simple numbers (1D tensors), let's deal with something more complicated than that, a multichannel image. Every colour image has 3 channels (<font color="#ff0000">Red</font>, <font color="#00ff00">Green</font> and <font color="#0000ff">Blue</font>). Every channel has a matrix of numbers showing the intensity of a pixel at that particular position.

### Importing the libraries
We'll first start by importing the tensorflow library and other libraries that we'll use for this tutorial.

```python
import numpy as np                # For the maths (as a side library, in case we need it)
import tensorflow as tf           # Main tensorflow library
import matplotlib.pyplot as plt   # For image handling
filename = "TensorFlow_logo.png"  # Image file name (tensorflow logo)
```

Let's read the image first

```python
image = plt.imread(filename)       # Read the image file
plt.imshow(image)                  # Put it in the plot canvas
test_r, test_c = 50, 50            # Any test point (row, column) pair
print("Image shape is : {ims}".format(ims=image.shape))
red, green, blue = image[test_r, test_c, 0], image[test_r, test_c, 1], image[test_r, test_c, 2]
print("Color at the test point is ({r}, {g}, {b})".format(r=red, g=green, b=blue))
plt.show()
```
